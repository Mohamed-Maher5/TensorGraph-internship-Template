```python
import pymongo
import random
```

```python
class UtteranceGenerator:

  def __init__(self, db):

    self.db = db
    self.categories = []
    self.items = []
    self.sizes_prices = []
    self.ingredients = []
    self.ingredients_names = []
    self.ingredients_values = []
    self.ingradiants_objects_names_values = []
    self.ingredients_values_prices = []
    self.ingredients_enforcedIngredients = []
    self.ingredients_objects_values_qualifiers = []
    self.files_num = 1
    self.sizes = {
      "Sm": "Small",
      "Md": "Medium",
      "Lg": "Large",
      "XL": "X Large",
      "8\"": "8 inch",
      "12\"": "12",
      "2Ltr": "2 Liters",
      "Size": "",
      "None": ""
    }
    self.enforced = {
      'Choose Your Sauce': 'Sauce',
      'Choose Your Toppings': 'Toppings',
    }

  def get_items_category(self):
    for item in self.db.menu_item.find({}):
      if(len(item['ingredients']) != 0):
        self.categories.append(item["category"])

  def get_items(self):
    for item in self.db.menu_item.find({}):
      if(len(item['ingredients']) != 0):
        self.items.append(item['name'])

  def get_items_sizes_prices(self):
    for item in self.db.menu_item.find({}):
      if(len(item['ingredients']) != 0):
        self.sizes_prices.append(item['sizePrices'])

  def get_items_ingredients(self):
    for item in self.db.menu_item.find({}):
      if(len(item['ingredients']) != 0):
        self.ingredients.append(item['ingredients'])

  def ingredients_preprossing(self):
    for i in range(len(self.ingredients)):
      for j in range(len(self.ingredients[i])):
        for k in range(len(self.ingredients[i][j]['values'])):
          if(self.ingredients[i][j]['values'][k]['sizePrices'] == 0):
            self.ingredients[i][j]['values'][k]['sizePrices'] = []
    return self.ingredients

  def get_names_for_all_ingredients_objects(self):
    for i in range(len(self.ingredients)):
      ingredient_objects_names = []
      for j in range(len(self.ingredients[i])):
        ingredient_objects_names.append(self.ingredients[i][j]['name'])
      self.ingredients_names.append(ingredient_objects_names)

  def get_values_for_all_ingredients_ojects(self):
    for i in range(len(self.ingredients)):
      ingredient_objects_values = []
      for j in range(len(self.ingredients[i])):
        names_of_object = []
        for k in range(len(self.ingredients[i][j]['values'])):
          names_of_object.append(self.ingredients[i][j]['values'][k]['name'])
        ingredient_objects_values.append(names_of_object)
      self.ingredients_values.append(ingredient_objects_values)

  def mapping_ingradiants_objects_names_values(self):
    for i in range(len(self.ingredients_names)):
      names_values_list = []
      for j in range(len(self.ingredients_names[i])):
        names_values_dic = {}
        names_values_dic[self.ingredients_names[i][j]] = self.ingredients_values[i][j]
        names_values_list.append(names_values_dic)
      self.ingradiants_objects_names_values.append(names_values_list)

  def get_prices_for_all_ingredients_ojects(self):
    for i in range(len(self.ingredients)):
      ingradiant_objects_values_prices = []
      for j in range(len(self.ingredients[i])):
        price_of_object = []
        for k in range(len(self.ingredients[i][j]['values'])):
          prices = []
          for l in range(len(self.sizes_prices[i])):
            if(len(self.ingredients[i][j]['values'][k]['sizePrices']) == 0):
              prices.append(0.0)
            elif(len(self.ingredients[i][j]['values'][k]['sizePrices']) < len(self.sizes_prices[i])):
              prices.append(self.ingredients[i][j]['values'][k]['sizePrices'][0]['price'])
            else:
              prices.append(self.ingredients[i][j]['values'][k]['sizePrices'][l]['price'])
          price_of_object.append(prices)
        ingradiant_objects_values_prices.append(price_of_object)
      self.ingredients_values_prices.append(ingradiant_objects_values_prices)

  def get_enforcedIngredients_for_all_ingredients_objects(self):
    for i in range(len(self.ingredients)):
      ingradiant_objects_enforcedIngredients_state = []
      for j in range(len(self.ingredients[i])):
        ingradiant_objects_enforcedIngredients_state.append(self.ingredients[i][j]['enforcedIngredients'])
      self.ingredients_enforcedIngredients.append(ingradiant_objects_enforcedIngredients_state)

  def get_values_qualifiers_for_all_ingredients_ojects(self):
    for i in range(len(self.ingredients)):
      ingredient_objects_values_qualifiers = []
      for j in range(len(self.ingredients[i])):
        values_objects_qualifier = []
        for k in range(len(self.ingredients[i][j]['values'])):
          object_qualifiers = []
          for l in range(len(self.ingredients[i][j]['values'][k]['qualifier'])):
            object_qualifiers.append(self.ingredients[i][j]['values'][k]['qualifier'][l]['name'])
          if(len(self.ingredients[i][j]['values'][k]['qualifier']) == 0):
            object_qualifiers.append("")
          values_objects_qualifier.append(object_qualifiers)
        ingredient_objects_values_qualifiers.append(values_objects_qualifier)
      self.ingredients_objects_values_qualifiers.append(ingredient_objects_values_qualifiers)

  def read_input_script(self):
    input_script = []
    file = open("input.txt", "r", encoding="UTF-8")
    content = file.read()
    file.close()
    input_script.append(content.split("\n"))
    return input_script

  def count_ingredients(self):
    var = 1
    target_string = "ingredient_value"+str(var)
    lines = self.read_input_script()[0]
    ingredients_num = 0
    for line in lines:
      for i in line:
        index = line.find(target_string)
        if index != -1:
          var+=1
          target_string = "ingredient_value"+str(var)
          ingredients_num +=1
    return ingredients_num

  def validate_and_update_input_script_inCase_qualifier(self,ingredients_num):
    var = 1
    input_script = self.read_input_script()[0]
    target_string = "{qualifier_value"+str(var)+"}"
    string_to_list = input_script[1].split(" ")
    if "{qualifier_value"+str(var)+"}" not in string_to_list:
      for i in range(2,len(input_script)):
        for j in range(ingredients_num):
          if "{ingredient_value"+str(j+1)+"}" in input_script[i]:
            input_script[i] = input_script[i].replace("{ingredient_value"+str(j+1)+"}","with {ingredient_value"+str(j+1)+"}")
    return input_script

  def mapping_sizes(self,size):
    if size in self.sizes:
      return self.sizes[size]
    else:
      return size

  def mapping_enforced_names(self,enforced_name):
    for i in range(len(enforced_name)):
      if enforced_name[i] in self.enforced:
        enforced_name[i] = self.enforced[enforced_name[i]]
    return enforced_name

  def get_category_indexes(self,item_category):
    category_indexes = []
    for i in range(len(self.categories)):
      if self.categories[i] == item_category:
        category_indexes.append(i)
    return category_indexes

  def generate_random_items_numbers(self,number_of_random_items,min_value,max_value):
    random_numbers = []
    for i in range(number_of_random_items):
      if(min_value == max_value):
        random_numbers.append(min_value)
      else:
        random_number = random.randint(min_value, max_value)
        while random_number in random_numbers and len(self.ingredients[random_number]) < self.count_ingredients():
          random_number = random.randint(min_value, max_value)
        random_numbers.append(random_number)
    return random_numbers

  def generate_random_ingredients_numbers(self,number_of_random_items,min_value,max_value,item_num):
    random_numbers = []
    for i in range(number_of_random_items):
      if(min_value == max_value):
        random_numbers.append(min_value)
      else:
        random_number = random.randint(min_value, max_value)
        while random_number in random_numbers or self.ingredients_enforcedIngredients[item_num][random_number] == 1:
          random_number = random.randint(min_value, max_value)
        random_numbers.append(random_number)
    return random_numbers

  def generate_random_numbers_for_category_items(self,number_of_random_items,min_value,max_value,category):
    random_numbers = []
    category_indexes = self.get_category_indexes(category)
    for i in range(number_of_random_items):
      if(min_value == max_value):
        random_numbers.append(min_value)
      else:
        random_number = random.randint(0, len(category_indexes)-1)
        while category_indexes[random_number] in random_numbers and len(self.ingredients[category_indexes[random_number]]) < self.count_ingredients():
          random_number = random.randint(min_value, len(category_indexes)-1)
        random_numbers.append(category_indexes[random_number])
    return random_numbers

  def generate_random_item_with_random_ingredients(self,random_item_num):
    ingredients_num = self.count_ingredients()
    random_choose = 0
    random_item_data = {}

    ingradiants_enforcedIngredients_names_equal_to_1 = []
    ingradiants_enforcedIngredients_values_equal_to_1 = []
    ingradiants_enforcedIngredients_values_qualifier_equal_to_1 = []

    random_item_ingradiant_object_num_list = []
    random_item_ingradiant_object_value_num_list = []
    random_item_ingradiant_object_values_list = []
    random_items_ingredients_objects_values_qualifiers_list = []

    random_item_ingradiant_object_num_list.append(self.generate_random_ingredients_numbers(ingredients_num , min_value = 0 , max_value = len(self.ingredients[random_item_num])-1,item_num = random_item_num))

    for i in range(len(self.ingredients_enforcedIngredients[random_item_num])):
      if(self.ingredients_enforcedIngredients[random_item_num][i] == 1):
        ingradiants_enforcedIngredients_names_equal_to_1.append(self.ingredients_names[random_item_num][i])
        random_ingradiants_enforcedIngredients_names_equal_to_1 = random.randint(0,len(self.ingredients_values[random_item_num][i])-1)
        ingradiants_enforcedIngredients_values_equal_to_1.append(self.ingredients_values[random_item_num][i][random_ingradiants_enforcedIngredients_names_equal_to_1])
        random_ingredient_object_for_enforcedIngredients_names_equal_to_1 = random.randint(0,len(self.ingredients_objects_values_qualifiers[random_item_num][i])-1)
        random_ingredient_object_qualifier_for_enforcedIngredients_names_equal_to_1 = random.randint(0,len(self.ingredients_objects_values_qualifiers[random_item_num][i][random_ingredient_object_for_enforcedIngredients_names_equal_to_1])-1)

        if self.ingredients[random_item_num][i]['values'][random_ingradiants_enforcedIngredients_names_equal_to_1]['isDefault'] == True:
          random_index = random.randint(0, 1)
          list_of_qualifiers = ["without",self.ingredients_objects_values_qualifiers[random_item_num][i][random_ingredient_object_for_enforcedIngredients_names_equal_to_1][random_ingredient_object_qualifier_for_enforcedIngredients_names_equal_to_1]]
          random_choose = list_of_qualifiers[random_index]
        else:
          random_choose = self.ingredients_objects_values_qualifiers[random_item_num][i][random_ingredient_object_for_enforcedIngredients_names_equal_to_1][random_ingredient_object_qualifier_for_enforcedIngredients_names_equal_to_1]
        ingradiants_enforcedIngredients_values_qualifier_equal_to_1.append(random_choose)

    for i in range(ingredients_num):
      random_item_ingradiant_object_value_num = random.randint(0,len(self.ingredients_values[random_item_num][random_item_ingradiant_object_num_list[0][i]])-1)
      random_item_ingradiant_object_value_num_list.append(random_item_ingradiant_object_value_num)
      random_item_ingradiant_object_values_list.append(list(self.ingradiants_objects_names_values[random_item_num][random_item_ingradiant_object_num_list[0][i]].values())[0][random_item_ingradiant_object_value_num])
      random_items_ingredients_objects_values_qualifiers_num = random.randint(0,len(self.ingredients_objects_values_qualifiers[random_item_num][random_item_ingradiant_object_num_list[0][i]][random_item_ingradiant_object_value_num])-1)
      random_items_ingredients_objects_values_qualifiers_list.append(self.ingredients_objects_values_qualifiers[random_item_num][random_item_ingradiant_object_num_list[0][i]][random_item_ingradiant_object_value_num][random_items_ingredients_objects_values_qualifiers_num])

    total_cost = 0.0

    random_item_size_price = random.randint(0,len(self.sizes_prices[random_item_num])-1)

    for i in range(ingredients_num):
      total_cost+=self.ingredients_values_prices[random_item_num][random_item_ingradiant_object_num_list[i][0]][random_item_ingradiant_object_value_num_list[i]][random_item_size_price]
    random_num_of_orders = 1

    item_name = self.items[random_item_num]
    item_category = self.categories[random_item_num]
    size = self.mapping_sizes(self.sizes_prices[random_item_num][random_item_size_price]['size'])
    quantity = random_num_of_orders
    total_cost += self.sizes_prices[random_item_num][random_item_size_price]['price']
    enforced_names = self.mapping_enforced_names(ingradiants_enforcedIngredients_names_equal_to_1)

    random_item_data.update({
        'item_name':item_name,
        'item_category':item_category,
        'ingredient_value':random_item_ingradiant_object_values_list,
        'qualifier_value':random_items_ingredients_objects_values_qualifiers_list,
        'size':str(size),
        'quantity':str(quantity),
        'total_cost':str(total_cost),
        'enforcedIngredient_name':enforced_names,
        'enforcedIngredient_value':ingradiants_enforcedIngredients_values_equal_to_1,
        'enforcedIngredient_value_qualifier':ingradiants_enforcedIngredients_values_qualifier_equal_to_1
        })
    return random_item_data

  def appendLines_incase_enforcedIngredient_state_equal_1(self,random_item_data):
    ingredients_num = self.count_ingredients()
    input_script = self.validate_and_update_input_script_inCase_qualifier(ingredients_num)
    for i in range(len(random_item_data['enforcedIngredient_name'])):
      input_script.insert(len(input_script)-2,"#bot")
      input_script.insert(len(input_script)-2,"Which {enforcedIngredient_name"+str(i+1)+"} do you want to add to your {item}?")
      input_script.insert(len(input_script)-2,"#me")
      input_script.insert(len(input_script)-2,"{enforcedIngredient_value"+str(i+1)+"}")
      input_script.pop()

      greate_message = "Great! {quantity} {item} "
      for i in range(len(random_item_data['enforcedIngredient_name'])):
        if(random_item_data['enforcedIngredient_name'][i] == ''):
          greate_message+="{enforcedIngredient_value_qualifier"+str(i+1)+"} {enforcedIngredient_value"+str(i+1)+"}"
        else:
          greate_message+="with {enforcedIngredient_value"+str(i+1)+"}"

        if(len(random_item_data['enforcedIngredient_name']) > 1 and i < len(random_item_data['enforcedIngredient_name'])-1):
          greate_message+=","

      for i in range(len(random_item_data['ingredient_value'])):
        if(random_item_data['qualifier_value'][i] == ''):
          greate_message+=" ,with {ingredient_value"+str(i+1)+"}"
        else:
          greate_message+=", {qualifier_value"+str(i+1)+"} {ingredient_value"+str(i+1)+"}"
        if((len(random_item_data['ingredient_value']) > 1 and i < len(random_item_data['ingredient_value'])-1) or (len(random_item_data['enforcedIngredient_name']) > 1 and i < len(random_item_data['ingredient_value'])-1)):
          greate_message+=","

      greate_message+=", and total cost is {cost}. Would you like me to add anything else?"
      input_script.append(greate_message)
    return input_script

  def mapping_random_item_data_with_replacements_keywords(self,random_item_data):
    replacements = {'item': random_item_data['item_name'],
                    'category': random_item_data['item_category'],
                    'size': random_item_data['size'],
                    'quantity': random_item_data['quantity'],
                    'cost':random_item_data['total_cost'],
                    }
    for i in range(len(random_item_data['ingredient_value'])):
      replacements.update({
          'ingredient_value'+str(i+1) : random_item_data['ingredient_value'][i]
      })

    for i in range(len(random_item_data['enforcedIngredient_name'])):
      replacements.update({
          'enforcedIngredient_name'+str(i+1) : random_item_data['enforcedIngredient_name'][i]
      })

    for i in range(len(random_item_data['enforcedIngredient_value'])):
      replacements.update({
          'enforcedIngredient_value'+str(i+1) : random_item_data['enforcedIngredient_value'][i]
      })

    for i in range(len(random_item_data['qualifier_value'])):
      if(random_item_data['qualifier_value'][i] == ''):
        random_item_data['qualifier_value'][i] = 'with'
      replacements.update({
          'qualifier_value'+str(i+1) : random_item_data['qualifier_value'][i]
      })
    for i in range(len(random_item_data['enforcedIngredient_value_qualifier'])):
      replacements.update({
          'enforcedIngredient_value_qualifier'+str(i+1) : random_item_data['enforcedIngredient_value_qualifier'][i]
      })
    return replacements

  def replace_keys_with_replacements(self,input_script,replacements):
    for i in range(len(input_script)):
        response = input_script[i]
        for key, value in replacements.items():
          response = response.replace('{' + key + '}', value)
        input_script[i] = response
    return input_script

  def write_script(self,new_script):
    with open("output_folder/output"+str(self.files_num)+".txt", "w") as file:
      for i in range(len(new_script)):
        file.write(new_script[i]+"\n")
    self.files_num += 1

  def generate_new_script(self,num_of_iteration,category = "none"):
    random_nums = []
    if category == "none" :
      random_nums = self.generate_random_items_numbers(num_of_iteration,0,len(self.items)-1)
    else:
       random_nums = self.generate_random_numbers_for_category_items(num_of_iteration,0,len(self.items)-1,category = category)
    for i in range(num_of_iteration):
      random_item_data = self.generate_random_item_with_random_ingredients(random_nums[i])
      input_script = self.appendLines_incase_enforcedIngredient_state_equal_1(random_item_data)
      replacements = self.mapping_random_item_data_with_replacements_keywords(random_item_data)
      new_script = self.replace_keys_with_replacements(input_script,replacements)
      self.write_script(new_script)
    self.files_num = 0

  def run(self):
    self.get_items_category()
    self.get_items()
    self.get_items_sizes_prices()
    self.get_items_ingredients()
    self.ingredients_preprossing()
    self.ingredients_preprossing()
    self.get_names_for_all_ingredients_objects()
    self.get_values_for_all_ingredients_ojects()
    self.mapping_ingradiants_objects_names_values()
    self.get_prices_for_all_ingredients_ojects()
    self.get_enforcedIngredients_for_all_ingredients_objects()
    self.get_values_qualifiers_for_all_ingredients_ojects()
```
